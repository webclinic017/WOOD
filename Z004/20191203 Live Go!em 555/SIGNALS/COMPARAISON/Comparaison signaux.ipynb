# COMPARAISON AVEC LE SYSTEME NEW_BT

```python
df1 = pd.read_csv('Signals_11_29.csv')
df1 = df1.drop(['Unnamed: 0','Open.S Price'],axis=1)
df1 = df1.sort_values('Ticker')
df1 = df1.drop_duplicates()
df1 = df1[['Date','Ticker','Open Long','Close Long','Open Short','Close Short','Close Price']]
df2 = pd.read_csv('SIGNALS_2019-11-29.csv')
df2 = df2.drop(['Unnamed: 0'],axis=1)
df2 = df2.sort_values('Ticker')
df2 = df2.drop_duplicates()
df2 = df2[['Date','Ticker','Open Long','Close Long','Open Short','Close Short','Close Price']]
print('Longueur des df1 & df2 :', df1.shape[0],' & ',df2.shape[0],' = ',abs(df1.shape[0]-df2.shape[0]),len(pd.concat([df1, df2]).loc[df1.index.symmetric_difference(df2.index)]))
pd.concat([df1, df2]).loc[df1.index.symmetric_difference(df2.index)]
```
