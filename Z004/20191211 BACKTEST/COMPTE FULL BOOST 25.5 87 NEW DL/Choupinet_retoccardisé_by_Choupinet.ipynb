```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import time
from colorama import Fore, Back, Style

tmps55=time.time()
df = pd.read_csv('SIGNALS_live.csv')

#df = df[df['Date'] >= '2020-01-14']

try:
    df = df.drop(['Unnamed: 0'],axis=1)
except:
    pass
choupinet = pd.DataFrame()

DT = []
OL = []
CL = []
OS = []
CS = []
XP = []
PNL = []
CUM = []
YMCA = []
NETL = []
NETS = []
LAT = []
XPL = []
XPS = []
MDD = []
NXP = []

ol = 0
cl = 0
os = 0
cs = 0
xp = 0
pnl = 0
cum = 0
ymca = 0
netl = 0
nets = 0
lat = 0
xpl = 0
xps = 0
mdd = 0
nxp = 0

for i in range(1,df.shape[0]):
    
    i = i-1
    
    long = df.shape[0]-1
    
    if i == 1:
            print('\r',Fore.GREEN,' |==',Fore.BLUE,'O',Fore.RED,'==================|',end='',flush=True)
    if i == round(1 * long / 9):
        print('\r',Fore.GREEN,' |====',Fore.BLUE,'O',Fore.RED,'================|',end='',flush=True)
    if i == round(2 * long / 9):
        print('\r',Fore.GREEN,' |======',Fore.BLUE,'O',Fore.RED,'==============|',end='',flush=True)
    if i == round(3 * long / 9):
        print('\r',Fore.GREEN,' |========',Fore.BLUE,'O',Fore.RED,'============|',end='',flush=True)
    if i == round(4 * long / 9):
        print('\r',Fore.GREEN,' |==========',Fore.BLUE,'O',Fore.RED,'==========|',end='',flush=True)
    if i == round(5 * long / 9):
        print('\r',Fore.GREEN,' |============',Fore.BLUE,'O',Fore.RED,'========|',end='',flush=True)
    if i == round(6 * long / 9):
        print('\r',Fore.GREEN,' |==============',Fore.BLUE,'O',Fore.RED,'======|',end='',flush=True)
    if i == round(7 * long / 9):
        print('\r',Fore.GREEN,' |================',Fore.BLUE,'O',Fore.RED,'====|',end='',flush=True)
    if i == round(8 * long / 9):
        print('\r',Fore.GREEN,' |==================',Fore.BLUE,'O',Fore.RED,'==|',Style.RESET_ALL,end='',flush=True)
    
    
    if df.iloc[i+1,0] == df.iloc[i,0]:
        ol += df.iloc[i]['Open Long']   #2
        cl += df.iloc[i]['Close Long']  #3
        os += df.iloc[i]['Open Short']  #4
        cs += df.iloc[i]['Close Short'] #5
        
        """if df.iloc[i-1,3] == 1 or df.iloc[i-1,5] == 1 :
            xp = xp - (df.iloc[i-1,6] * df.iloc[i-1,7])
        
        if df.iloc[i-1,2] == 1 or df.iloc[i-1,4] == 1 :
            xp = xp + (df.iloc[i-1,6] * df.iloc[i-1,7])
        
        try:
            cum += df.iloc[i-1,9]
        except:
            cum += 0
        pnl += df.iloc[i-1,9]
        lat += df.iloc[i-1,10]
        
        if df.iloc[i-1,2] == 1:
            xpl = xpl + (df.iloc[i-1,6] * df.iloc[i-1,7])
        if df.iloc[i-1,3] == 1:
            xpl = xpl - (df.iloc[i-1,6] * df.iloc[-1,7])
        
        if df.iloc[i-1,4] == 1:
            xps = xps  - (df.iloc[i-1,6] * df.iloc[i-1,7])
        if df.iloc[i-1,5] == 1:
            xps = xps + (df.iloc[i-1,6] * df.iloc[i-1,7])
        
        
        nxp =xpl+  xps"""
        
        if df.iloc[i]['Close Long'] == 1 or df.iloc[i]'Close Short'] == 1 :

            xp = xp - (df.iloc[i]['Buy/Sell Price'] * df.iloc[i]['Quantity'])

        if df.iloc[i,2] == 1 or df.iloc[i,4] == 1 :

            xp = xp + (df.iloc[i,6] * df.iloc[i,7])

        try:
            cum += df.iloc[i,9]
        except:
            cum += 0
        pnl += df.iloc[i,9]
        lat += df.iloc[i,10]

        if df.iloc[i,2] == 1:

            xpl = xpl + (df.iloc[i,6] * df.iloc[i,7])

        if df.iloc[i,3] == 1:

            xpl = xpl - (df.iloc[i,6] * df.iloc[i,7])

        if df.iloc[i,4] == 1:

            xps = xps  - (df.iloc[i,6] * df.iloc[i,7])

        if df.iloc[i,5] == 1:

            xps = xps + (df.iloc[i,6] * df.iloc[i,7])

        nxp = xpl + xps

        
    else:
        
        ol += df.iloc[i,2]
        cl += df.iloc[i,3]
        os += df.iloc[i,4]
        cs += df.iloc[i,5]
        
        """if df.iloc[i-1,3] == 1 or df.iloc[i-1,5] == 1 :
            xp = xp - (df.iloc[i-1,6] * df.iloc[i-1,7])
        if df.iloc[i-1,2] == 1 or df.iloc[i-1,4] == 1 :
            xp = xp + (df.iloc[i-1,6] * df.iloc[i-1,7])
        
        try:
            cum += df.iloc[i-1,9]
        except:
            cum += 0
        pnl += df.iloc[i-1,9]
        lat += df.iloc[i-1,10]
        
        if df.iloc[i-1,2] == 1:
            xpl = xpl + (df.iloc[i-1,6] * df.iloc[i-1,7])
        
        if df.iloc[i-1,3] == 1:
            xpl = xpl - (df.iloc[i-1,6] * df.iloc[i-1,7])
        
        if df.iloc[i-1,4] == 1:
            xps = xps  - (df.iloc[i-1,6] * df.iloc[i-1,7])
        
        if df.iloc[i-1,5] == 1:
            xps = xps + (df.iloc[i-1,6] * df.iloc[i-1,7])
        
        
        nxp =xpl+  xps"""
        
        if df.iloc[i,3] == 1 or df.iloc[i,5] == 1 :

            xp = xp - (df.iloc[i,6] * df.iloc[i,7])

        elif df.iloc[i,2] == 1 or df.iloc[i,4] == 1 :

            xp = xp + (df.iloc[i,6] * df.iloc[i,7])

        try:
            cum += df.iloc[i,9]
        except:
            cum += 0
        pnl += df.iloc[i,9]
        lat += df.iloc[i,10]

        if df.iloc[i,2] == 1:

            xpl = xpl + (df.iloc[i,6] * df.iloc[i,7])

        if df.iloc[i,3] == 1:

            xpl = xpl - (df.iloc[i,6] * df.iloc[i,7])

        if df.iloc[i,4] == 1:

            xps = xps  - (df.iloc[i,6] * df.iloc[i,7])

        if df.iloc[i,5] == 1:

            xps = xps + (df.iloc[i,6] * df.iloc[i,7])

        nxp = xpl + xps

        ymca = 1000000 + cum +lat
        XP.append(xp)
        PNL.append(pnl)
        CUM.append(cum)
        XPL.append(xpl)
        XPS.append(xps)
        NXP.append(nxp)
        LAT.append(lat)
        if min(LAT) < 0:
            mdd = min(LAT)
        MDD.append(mdd)
        YMCA.append(ymca)
        pnl = 0
        lat = 0
        mdd = 0
        
        netl = netl + ol - cl
        nets = nets + os - cs
        DT.append(df.iloc[i,0])
        OL.append(ol)
        CL.append(cl)
        OS.append(os)
        CS.append(cs)
        
        NETL.append(netl)
        NETS.append(nets)
        
        
        ol = 0
        cl = 0
        os = 0
        cs = 0
        


            
choupinet['Date'] = DT
choupinet['Open Long'] = OL
choupinet['Close Long'] = CL
choupinet['Open Short'] = OS
choupinet['Close Short'] = CS
choupinet['Exposure Long'] = XPL
choupinet['Exposure Short'] = XPS
choupinet['Exposure'] = XP
choupinet['Net Exposure'] = NXP
choupinet['PNL'] = PNL
choupinet['Unrealized PNL'] = LAT
choupinet['Cumulative PNL'] = CUM
choupinet['NAV'] = YMCA
choupinet['Net Long'] = NETL
choupinet['Net Short'] = NETS
choupinet['Maximum Drawdown'] = MDD

choupinet.to_csv('portfolio_live.csv')

plt.figure(figsize=(16,8),dpi=300)
plt.title('Equity curve du début de tout au 7 fevrier')
plt.ylabel('NAV')
plt.xlabel('Days')
plt.plot(choupinet['NAV'],color='purple',lw=0.9)
plt.savefig('nav_live.png')
tmps22=round(time.time()-tmps55,2)
print(Fore.BLUE,"/n Time for completing Choupinetisation Generation = %f" %tmps22,'seconds \n',Style.RESET_ALL)
choupinet.tail()
```
