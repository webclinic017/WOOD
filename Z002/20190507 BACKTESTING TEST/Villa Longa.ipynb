```python
print('Importation des librairies...')
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
print('Librairies importées')
```

```python
# SCRAPING DES DONNES BRUTES
df = pd.read_csv('gbpjpy_15mn_1Y.csv')
df = df.set_index('Date')
df = df.iloc[:,1:-1]
df.head()
```

```python
equity_long = 100000
equity_short = 100000
ticket_long = equity_long/2
ticket_short = equity_short/2
longueur = len(df)-1
long_pos = 0
short_pos = 0
target_long = 0.3
target_short = -0.35
sl_long = -1.9
sl_short = 1.9
long_on_price = 0
short_on_price = 0
nbre_positions = 0
profit_long = 0
profit_short = 0
long_trick = 0
short_trick = 0
pactol = 0
```

```python
print(' Pour le début du test, entrée forcée en position long et short le ',df.index[0])

long_on_price = df.iloc[0,3]
long_pos = 1
equity_long = equity_long - 1.5
ticket_long = ticket_long/2
#long_trick = 4
print(' Forçage Long sur une position à ',long_on_price,'\n')

short_on_price = df.iloc[1,3]
print(' Forçage Short sur une position à ',long_on_price,'\n')
short_pos = 1
equity_short = equity_short - 1.5
ticket_short = ticket_short/2
#short_trick = 4

print('_________________________________________________\n')

for i in range (20,longueur):
    
    if ticket_long > 1:
        pactol = pactol + equity_long - 100000
        equity_long =100000
        ticket_long = 50000
    
    if ticket_short < 1:
        pactol = pactol + equity_short - 100000
        equity_short = 100000
        ticket_short = 50000
    
    else:
        b = i-2
        pnl = df.iloc[i,3] - df.iloc[b,3]
        #short_trick = short_trick - 1
        #long_trick = long_trick - 1
        
        print('\nRang i ',i,' et pnl = ',pnl,'prix -2 :',df.iloc[b,3],'prix actuel: ',df.iloc[i,3], '\n')

        if equity_long < 0:
            long_pos = 2
            
        elif equity_short < 0:
            short_pos = 2
        
        if pnl > target_long and long_pos == 1 and long_trick != 4:
            long_pos = 0
            #long_trick = 4
            profit_long = pnl*ticket_long
            equity_long = equity_long - 1.5 + profit_long
            nbre_positions += 1
            print(' Close du long ',df.iloc[i,3],' le ',df.index[i])
            print(" Le prix d'achat était de ",long_on_price)
            print(' Le ticket était de ',ticket_long)
            print('Variation : ',pnl)
            print('pnl ',profit_long)
            print('Nouveau ticket : ',ticket_long)
            print('Nouvelle equity Long ',equity_long,'\n')
            ticket_long = ticket_long/2
            


        elif pnl < sl_long and long_pos == 1:
            long_pos = 0
            #long_trick = 4
            profit_long = pnl*ticket_long
            equity_long = equity_long - 1.5 + profit_long
            nbre_positions += 1
            print(' Close du long sur stop loss à ',df.iloc[i,3],' le ',df.index[i])
            print(" Le prix d'achat était de ",long_on_price)
            print(' Le ticket était de ',ticket_long)
            print('Variation : ',pnl)
            print('pnl ',profit_long)
            print('Nouvelle equity Long ',equity_long,'\n')
            print('Nouveau ticket : ',ticket_long)
            ticket_long = ticket_long/2

        elif pnl < target_short and short_pos == 1 and short_trick != 4:
            short_pos = 0
            #short_trick = 4
            profit_short = abs(pnl*ticket_short)
            equity_short = equity_short - 1.5 + profit_short
            nbre_positions += 1
            print(' Close du long ',df.iloc[i,3],' le ',df.index[i])
            print(" Le prix d'achat était de ",short_on_price)
            print(' Le ticket était de ',ticket_short)
            print('Variation : ',abs(pnl))
            print('pnl ',profit_short)
            print('Nouveau ticket : ',ticket_long)
            print('Nouvelle equity Long ',equity_short,'\n')
            ticket_short = ticket_short/2


        elif pnl > sl_short and short_pos == 1:
            short_pos = 0
            #short_trick = 4
            profit_short = -(pnl*ticket_short)
            equity_short = equity_short - 1.5 + profit_short
            nbre_positions += 1
            print(' Close du long sur stop loss à ',df.iloc[i,3],' le ',df.index[i])
            print(" Le prix d'achat était de ",short_on_price)
            print(' Le ticket était de ',ticket_short)
            print('Variation : ',-pnl)
            print('pnl ',profit_short)
            print('Nouveau ticket : ',ticket_long)
            print('Nouvelle equity Long ',equity_short,'\n')
            ticket_short = ticket_short/2

        elif long_pos == 0 :#and long_trick == 0:
            long_on_price = df.iloc[i,3]
            long_pos = 1
            equity_long = equity_long - 1.5
            print(' Long on sur une position à ',long_on_price,'\n')



        elif short_pos == 0 :#and short_trick == 0:
            short_on_price = df.iloc[i,3]
            print(' Short on sur une position à ',long_on_price,'\n')
            short_pos = 1
            equity_short = equity_short - 1.5
            print('Nouvelle equity Short ',equity_short,'\n')
            
```

```python
print('Equity Long finale',equity_long)
print('Gain en long :',equity_long-100000)
print('Equity Short finale',equity_short)
print('Gain en short :',equity_short-100000)
print('Nbre positions',nbre_positions)
print(' Pactol :',pactol)
print('gain total :',round(pactol+equity_short+equity_long-200000))
```
