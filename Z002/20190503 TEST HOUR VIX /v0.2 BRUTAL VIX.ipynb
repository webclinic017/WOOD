```python
spy_close = []
oc = []
hl = []
spy_oc = []
spy_hl = []
spy_volume = []
vix_close = []
```

```python
for i in range (0,len_spy):
    oc.append(float(spy.iloc[i,0]) - float(spy.iloc[i,3]))
    hl.append(float(spy.iloc[i,1]) - float(spy.iloc[i,2]))
```

```python
len(oc),len(hl),len(spy)
```

```python
spy['oc'] = oc
spy['hl'] = hl
spy.head()
```

```python
for i in range (1,len_spy):
    b = i - 1
    spy_close.append((spy.iloc[i,3] - spy.iloc[b,3]) / (spy.iloc[b,3]+0.0001))
    spy_volume.append((spy.iloc[i,4] - spy.iloc[b,4]) / (spy.iloc[b,4]+0.0001))
    spy_oc.append((spy.iloc[i,5] - spy.iloc[b,5]) / (spy.iloc[b,5]+0.0001))
    spy_hl.append((spy.iloc[i,6] - spy.iloc[b,6]) / (spy.iloc[b,6]+0.0001))
```

```python
for i in range (1,len_vix): 
    b = i - 1
    vix_close.append((vix.iloc[i,-1] - vix.iloc[b,-1]) / vix.iloc[b,-1])
```

```python
spy = spy.iloc[1:,:]
vix = vix.iloc[1:,:]
```

```python
len(spy),len(vix),len(spy_close),len(vix_close),len(spy_oc),len(spy_volume),len(spy_hl)
```

```python
spy['spy_oc'] = spy_oc
spy['spy_hl'] = spy_hl
spy['spy_volume'] = spy_volume
spy['spy_close'] = spy_close
```

```python
vix['vix_close'] = vix_close
```

```python
df_raw = pd.merge(left=spy.iloc[1:,:],right=vix.iloc[1:,:],left_index=True,right_index=True)
```
