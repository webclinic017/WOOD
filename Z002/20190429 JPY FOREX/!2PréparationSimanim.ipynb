```python
macd, macdsignal, macdhist = talib.MACD(df['Close_eurusd'], fastperiod=50, slowperiod=200, signalperiod=20)
```

```python
rsi = talib.RSI(df['Close_eurusd'], timeperiod=5)
```

```python
upperband, middleband, lowerband = talib.BBANDS(df['Close_eurusd'], timeperiod=20, nbdevup=2, nbdevdn=2, matype=0)
```

```python
atr = talib.ATR(df['High_eurusd'], df['Low_eurusd'], df['Close_eurusd'], timeperiod=14)
```

```python
wclprice = talib.WCLPRICE(df['High_eurusd'], df['Low_eurusd'], df['Close_eurusd'])
```

```python
df['Macdsignal_eurusd'] = macdsignal
df['rsi_eurusd'] = rsi
df['BBup_eurusd'] = upperband
df['BBmid_eurusd'] = middleband
df['BBlow_eurusd'] = lowerband
df['atr_eurusd'] = atr
df['wclprince_eurusd'] = wclprice
df.head()
```
