```python
print('Importing librairies...')
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from sklearn.preprocessing import MinMaxScaler
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_auc_score,roc_curve
from keras.models import Sequential
from keras.layers import Input
from keras.layers import Dense
from keras.utils import to_categorical
%matplotlib inline
print('Librairies imported')
```

```python
tf.logging.set_verbosity(tf.logging.ERROR)
```

```python
df_1D = pd.read_csv('20190526 data prepared.csv')
df_1D = df_1D.iloc[:,1:]
df_1D = df_1D.dropna()
df_1D.reset_index(inplace=True,drop=True)
```

```python
delta = round(df_1D.shape[0])
bloc1 = round(delta*0.8)
bloc2 = delta - bloc1
print("# of periods : ",delta)
print('On 80% - 20% slash : ')
print('Bloc 1 : ',bloc1,' periods (',df_1D.iloc[bloc1,0],')\nBloc 2 :',bloc2,' periods (',df_1D.iloc[bloc2,0],')')
print('First period :',df_1D.iloc[0,0])
print('Last period :',df_1D.iloc[df_1D.shape[0]-1,0])
```

```python
X = df_1D.iloc[:,1:-4]
#X = X.drop(['Peaks','Valley'],axis=1)
y_up = to_categorical(df_1D.iloc[:,-3:-2])
y_down = to_categorical(df_1D.iloc[:,-2:-1])
```

```python
scaler = MinMaxScaler(feature_range=(-1, 1))

X = scaler.fit_transform(X)
y_up = scaler.fit_transform(y_up)
y_down = scaler.fit_transform(y_down)
```

```python
Xtrain = X[:bloc1,:]
Xtest = X[bloc1:,:]
ytrain_up = y_up[:bloc1,:]
ytest_up = y_up[bloc1:,:]
ytrain_down = y_down[:bloc1,:]
ytest_down = y_down[bloc1:,:]
```

```python
seed = 770
np.random.seed(seed)
# create model
model = Sequential()
model.add(Dense(50, activation='relu', input_dim=22))
model.add(Dense(1, activation='softmax'))
# Compile model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(x=Xtrain,
    y=ytrain_up,
    batch_size=32,
    epochs=300,
    verbose=1,
    callbacks=None,
    validation_split=0.2,
    validation_data=None,
    shuffle=True,
    class_weight=None,
    sample_weight=None,
    initial_epoch=0,
    steps_per_epoch=None,
    validation_steps=None,
)
```

```python
model.test_on_batch(Xtest, ytest)
model.metrics_names
```

```python
plt.figure(figsize=(26,6))
plt.plot(history.history['loss'])
plt.plot(history.history['acc'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()
```

```python
model.save('mlp.h5')
```

```python
print('Initializing \b Processing')
yhat = model.predict(Xtest, verbose=1)
print('All processed')
```
