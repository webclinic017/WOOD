```python
import pandas as pd
import os
import shutil
from bs4 import BeautifulSoup
import time
from colorama import Fore, Back, Style
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

os.system('say "Okay pticrapaud. On commence! Youpihihihi! Espèssdecafard"')
___Author___='LumberJack Jyss'
print('Global Optimized LumberJack Environment Motor 55\nLumberJack Jyss 5779(c)')
print(Fore.BLUE,'°0Oo_D.A.G._26_oO0°')
print(Fore.YELLOW,Back.BLUE,'Reporting v0.55',Style.RESET_ALL)

list_dir = os.listdir('.')

if str(list_dir).split(',')[0].split('.')[1].split("'")[0]=='xls':
    compte = str(list_dir).split(',')[0].split('_')[0][-2:]
elif str(list_dir).split(',')[1].split('.')[1].split("'")[0]=='xls':
    compte = str(list_dir).split(',')[1].split('_')[0][-2:]
elif str(list_dir).split(',')[2].split('.')[1].split("'")[0]=='xls':
    compte = str(list_dir).split(',')[2].split('_')[0][-2:]
elif str(list_dir).split(',')[3].split('.')[1].split("'")[0]=='xls':
    compte = str(list_dir).split(',')[3].split('_')[0][-2:]
elif str(list_dir).split(',')[4].split('.')[1].split("'")[0]=='xls':
    compte = str(list_dir).split(',')[4].split('_')[0][-2:]
    
try:
    os.mkdir('DONE/')
except:
    pass

try:
    report = pd.read_csv('report_'+compte+'.csv')
    DATE = report['Date'].tolist()  
    NAV_DEB = report['Starting NAV'].tolist()  
    NAV_FIN = report['Ending NAV'].tolist()  
    DIFF = report['Daily PNL in $'].tolist()
    PER = report['Daily PNL in %'].tolist()
except:
    
    DATE = []
    NAV_DEB = []
    NAV_FIN = []
    DIFF = []
    PER = []


long = len(list_dir)
tmps55=time.time()
for i,j in enumerate(sorted(list_dir)):
    if i == 1:
        print('\r',Fore.GREEN,' |==',Fore.BLUE,'O',Fore.RED,'==================|',end='',flush=True)
    if i == round(long/9):
        print('\r',Fore.GREEN,' |====',Fore.BLUE,'O',Fore.RED,'================|',end='',flush=True)
    if i == round(long/9)*2:
        print('\r',Fore.GREEN,' |======',Fore.BLUE,'O',Fore.RED,'==============|',end='',flush=True)
    if i == round(long/9)*3:
        print('\r',Fore.GREEN,' |========',Fore.BLUE,'O',Fore.RED,'============|',end='',flush=True)
    if i == round(long/9)*4:
        print('\r',Fore.GREEN,' |==========',Fore.BLUE,'O',Fore.RED,'==========|',end='',flush=True)
    if i == round(long/9)*5:
        print('\r',Fore.GREEN,' |============',Fore.BLUE,'O',Fore.RED,'========|',end='',flush=True)
    if i == round(long/9)*6:
        print('\r',Fore.GREEN,' |==============',Fore.BLUE,'O',Fore.RED,'======|',end='',flush=True)
    if i == round(long/9)*7:
        print('\r',Fore.GREEN,' |================',Fore.BLUE,'O',Fore.RED,'====|',end='',flush=True)
    if i == round(long/9)*8:
        print('\r',Fore.GREEN,' |==================',Fore.BLUE,'O',Fore.RED,'==|',end='',flush=True)
    
    if j != 'DONE' and j.split('.')[1] == 'xls' and j.split('.')[0].split('_')[1] == j.split('.')[0].split('_')[2]:
    
        try:
            table = BeautifulSoup(open(j,'r').read())
            df = pd.read_html(str(table))
        except:
            continue
        DATE.append(j.split('.')[0].split('_')[1][:4]+'-'+j.split('.')[0].split('_')[1][4:6]+'-'+j.split('.')[0].split('_')[1][6:])
        NAV_DEB.append(df[2].iloc[0,1]) 
        NAV_FIN.append(df[2][df[2]['Change in NAV'] == 'Ending Value'].iloc[0,1])
        DIFF.append(round((df[2][df[2]['Change in NAV'] == 'Ending Value'].iloc[0,1] - (df[2].iloc[0,1])),2))
        PER.append(round(((df[2][df[2]['Change in NAV'] == 'Ending Value'].iloc[0,1] - (df[2].iloc[0,1]))/(df[2].iloc[0,1]))*100,2))
        shutil.move(j, 'DONE/'+j)
tmps22=round(time.time()-tmps55,2)
print(Fore.BLUE,"\n Time for complete Generation = %f" %tmps22,'seconds \n',Style.RESET_ALL)

name = df[df=='Account Information'].iloc[0,1]
account = df[df=='Account Information'].iloc[1,1]
#df.title(account, ' ', name)
report = pd.DataFrame()
report['Date'] = DATE
report['Starting NAV'] = NAV_DEB
report['Ending NAV'] = NAV_FIN
report['Daily PNL in $'] = DIFF
report['Daily PNL in %'] = PER
report.to_csv('report_'+compte+'.csv')

mean = round((report.iloc[:]['Daily PNL in %'].sum())/len(report),5)
plt.figure(figsize=(20,10))
plt.plot(report.iloc[:,-1],label='Daily P&L in %',c='blue')
plt.axhline(y=mean,label='Mean of % daily p&l',c='orange')
plt.suptitle(name+' Account '+account,c='purple')
plt.title('Mean daily p&l = '+str(mean)+'%',c='red')
plt.legend()
plt.savefig('Report_'+compte+'.jpg')
report
os.system('say "Okay pticrapaud. Programme terminé. Espèssdecafard kanmèm"')
```
