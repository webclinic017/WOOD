```python
import pandas as pd
SYMBOL=[]
DATE=[]
QTY=[]
trade = pd.DataFrame()
final = pd.DataFrame()
```

```python
anterior = pd.read_csv('ANTERIOR_BASE.csv')
```

```python
ANTERIOR = anterior['Symbol'].to_list()
```

```python
traded = pd.read_csv('QTY_TRAITE_NOVEMBER.csv')
```

```python
traded = traded.sort_values('ReportDate')
```

```python
DATES = list(traded['ReportDate'].unique())
DATES
```

```python
TICKER = list(traded['Symbol'].unique())
len(TICKER)
```

```python
for x in ANTERIOR:
    TICKER.remove(x)
len(TICKER)
```

```python
for date in DATES:  
    for ticker in TICKER:
        tempo = traded[traded['ReportDate']==date] 
        tekko = tempo[tempo['Symbol']==ticker]
        SYMBOL.append(ticker)
        DATE.append(date)
        QTY.append(tekko['Quantity'].sum())
```

```python
trade['Symbol'] = SYMBOL
trade['Date'] = DATE
trade['Quantity'] = QTY
trade
```

```python
SYMBOL = []
DATE = []
OL = []
CL = []
OS = []
CS = []
poz = 0

for foo in trade['Symbol']:
    element = trade[(trade['Symbol']==foo)&(trade['Quantity']!=0)]
    for bar in range (0,len(element)):
    
        
        if poz == 0:
            poz = 1
            if element.iloc[bar]['Quantity'] > 0:
                OL.append(1)
                CL.append(0)
                OS.append(0)
                CS.append(0)
                DATE.append(element.iloc[bar]['Date'])
                SYMBOL.append(element.iloc[bar]['Symbol'])
            elif element.iloc[bar]['Quantity'] < 0:
                OL.append(0)
                CL.append(0)
                OS.append(1)
                CS.append(0)
                DATE.append(element.iloc[bar]['Date'])
                SYMBOL.append(element.iloc[bar]['Symbol'])
        elif poz == 1:
            poz = 0
            if element.iloc[bar]['Quantity'] > 0:
                OL.append(0)
                CL.append(0)
                OS.append(0)
                CS.append(1)
                DATE.append(element.iloc[bar]['Date'])
                SYMBOL.append(element.iloc[bar]['Symbol'])
            elif element.iloc[bar]['Quantity'] < 0:
                OL.append(0)
                CL.append(1)
                OS.append(0)
                CS.append(0)
                DATE.append(element.iloc[bar]['Date'])
                SYMBOL.append(element.iloc[bar]['Symbol'])
final['Date'] = DATE
final['Symbol'] = SYMBOL
final['Open Long'] = OL
final['Close Long'] = CL
final['Open Short'] = OS
final['Close Short'] = CS
final
```
