```python
import pandas as pd
SYMBOL=[]
DATE=[]
QTY=[]
trade = pd.DataFrame()
final = pd.DataFrame()

anterior = pd.read_csv('ANTERIOR_BASE.csv')

ANTERIOR = anterior['Symbol'].to_list()

traded = pd.read_csv('QTY_TRAITE_NOVEMBER.csv')

traded = traded.sort_values('ReportDate')

DATES = list(traded['ReportDate'].unique())
DATES
```

```python
TICKER = list(traded['Symbol'].unique())
len(TICKER)
```

```python
for x in ANTERIOR:
    TICKER.remove(x)
len(TICKER)
```

```python
for date in DATES:  
    for ticker in TICKER:
        tempo = traded[traded['ReportDate']==date] 
        tekko = tempo[tempo['Symbol']==ticker]
        SYMBOL.append(ticker)
        DATE.append(date)
        QTY.append(tekko['Quantity'].sum())
```

```python
trade['Symbol'] = SYMBOL
trade['Date'] = DATE
trade['Quantity'] = QTY
trade
```

```python
SYMBOL = []
DATE = []
OL = []
CL = []
OS = []
CS = []


for foo in trade['Symbol']:
    poz = 0
    element = trade[(trade['Symbol']==foo)&(trade['Quantity']!=0)]
    for bar in range (0,len(element)):
        if poz == 0:
            poz = 1
            if element.iloc[bar]['Quantity'] > 0:
                OL.append(1)
                CL.append(0)
                OS.append(0)
                CS.append(0)
                DATE.append(element.iloc[bar]['Date'])
                SYMBOL.append(element.iloc[bar]['Symbol'])
            if element.iloc[bar]['Quantity'] < 0:
                OL.append(0)
                CL.append(0)
                OS.append(1)
                CS.append(0)
                DATE.append(element.iloc[bar]['Date'])
                SYMBOL.append(element.iloc[bar]['Symbol'])
        else:
            poz = 0
            if element.iloc[bar]['Quantity'] > 0:
                OL.append(0)
                CL.append(0)
                OS.append(0)
                CS.append(1)
                DATE.append(element.iloc[bar]['Date'])
                SYMBOL.append(element.iloc[bar]['Symbol'])
            if element.iloc[bar]['Quantity'] < 0:
                OL.append(0)
                CL.append(1)
                OS.append(0)
                CS.append(0)
                DATE.append(element.iloc[bar]['Date'])
                SYMBOL.append(element.iloc[bar]['Symbol'])
final['Date'] = DATE
final['Symbol'] = SYMBOL
final['Open Long'] = OL
final['Close Long'] = CL
final['Open Short'] = OS
final['Close Short'] = CS
final = final.drop_duplicates()
```

```python
DATE = []
OL = []
CL = []
OS = []
CS = []
report = pd.DataFrame()
for date in DATES:
    DATE.append(date)
    OL.append(final[final['Date']==date]['Open Long'].sum())
    CL.append(final[final['Date']==date]['Close Long'].sum())
    OS.append(final[final['Date']==date]['Open Short'].sum())
    CS.append(final[final['Date']==date]['Close Short'].sum())
report['Date'] = DATE
report['Open Long'] = OL
report['Close Long'] = CL
report['Open Short'] = OS
report['Close Short'] = CS
```
