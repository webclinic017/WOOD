```python
import random
import pandas_datareader as web
#import quandl
from scipy.optimize import minimize
from scipy import stats
import matplotlib.pyplot as plt
%matplotlib inline
#import seaborn as sns
#sns.set(palette='bright',style='whitegrid')
```

```python
start = '2019-01-01'
end = '2019-03-18'
df = web.DataReader('KO','yahoo',start,end)
value = df.iloc[0,5]
df.head()
print("Modèle Bollinger Stochastic Descent de Cliff Benson #2")
mise = 450000
print('Mise de départ de ',mise,'. Valeur à ', value)

pool_long = mise/4
pool_short = mise/4
stock_long = (mise/4)/value
stock_short = (mise/4)/value
quot = 10
gain = 0
cumul = 0
ratio = 0
p = -1
lenght = len(df['Adj Close'])
print('________________')
print("Initialization")
print('________________')
for boucle in range(lenght) :
    n=value-df.iloc[boucle,5]
    print('n ',n)
    print('value ',value)
    value=df.iloc[boucle,5]
    p = p + 1
    if n == 0:
            print ('Phase',p)
            print("Valeur d'achat : ", value)
            print ('Pool_long : ',pool_long)
            print ('stock_long : ',stock_long)
            print ('pool_short : ',pool_short)
            print ('stock_short : ',stock_short)
            print('___________________________')
    
    elif n < 0 and pool_long > 0:
        print ('Phase',p)
        print("Valeur d'achat : ", value)
        print('Long on de ',((pool_long/quot) / value),' actions')
        print('Dépense du pool long de ',(pool_long/quot))
        stock_long = stock_long + ((pool_long/quot) / value)
        pool_long = pool_long - (pool_long/quot)
        print ('Pool_long : ',pool_long)
        print ('stock_long : ',stock_long)
        print ('pool_short : ',pool_short)
        print ('stock_short : ',stock_short)
        print('___________________________')
    
    elif n > 0 and stock_short > 1 :
            print ('Phase',p)
            print("Valeur d'achat : ", value)
            print('Short de ',(stock_short/quot),' actions')
            print('Gain du pool short de ',((stock_short/quot) * value))
            pool_short = pool_short + ((stock_short/quot) * value)
            stock_short = stock_short - (stock_short/quot)
            print ('Pool_long : ',pool_long)
            print ('stock_long : ',stock_long)
            print ('pool_short : ',pool_short)
            print ('stock_short : ',stock_short)
            print('___________________________')
            
    else :
        print ('Phase',p)
        print("Valeur d'achat : ", value)
        print ('Pool_long : ',pool_long)
        print ('stock_long : ',stock_long)
        print ('pool_short : ',pool_short)
        print ('stock_short : ',stock_short)
        print('___________________________')
        
print('___________________________________________________________________')
print("Il n'y a plus de fond à dépenser. Termination")
print('___________________________________________________________________')


print ('Phase',p)
print("Valeur d'achat : ", value)
print ('Pool_long : ',pool_long)
print ('stock_long : ',stock_long)
print ('pool_short : ',pool_short)
print ('stock_short : ',stock_short)
print('___________________________')



pool_short = pool_short + (stock_short * value)
stock_short = 0
pool_long = pool_long + (stock_long * value)
stock_long = 0

print('___________________________________________________________________')
print("Il n'y a plus de stock à vendre. Finalization")
print('___________________________________________________________________')

print ('Phase finale')
print("Valeur d'achat : ", value)
print ('Pool_long : ',pool_long)
print ('stock_long : ',stock_long)
print ('pool_short : ',pool_short)
print ('stock_short : ',stock_short)
print('___________________________')

cumul = pool_short + pool_long
print('le cumul est de : ',cumul)
gain = cumul - 400000
print ("Le gain total est de : ",gain)
ratio = gain/4000
print('Le ratio est de ',ratio,'%')
```
