```python
 from statsmodels.tsa.stattools import adfuller
```

```python
result = adfuller(df['Close_GSPC'].dropna())
```

```python
def adf_check (time_series):
    result=adfuller(time_series)
    print('Augmented Dicky-Fuller Test : ')
    labels=['ADF Test Statistic', 'p-value','# of lags','# of observations used']
    for value,label in zip(result,labels):
        print(label+' : '+str(value))
    if result[1] <= 0.05:
        print ('Strong evidence again null hypothesis')
        print('Reject null hypothesis')
        print ('Data has no unit root and is **stationary**')
    else:
        print ('Weak evidence again null hypothesis')
        print('Failed to reject null hypothesis')
        print ('Data has a unit root and is ** non-stationary**')
        
```

```python
adf_check(df['Close_GSPC'].dropna())
```

```python
df['First Difference']=df['Close_GSPC']-df['Close_GSPC'].shift(1)
```

```python
 adf_check(df['First Difference'].dropna())
```

```python
df['Second Difference']=df['First Difference']-df['First Difference'].shift(1)
```

```python
 adf_check(df['Second Difference'].dropna())
```

```python
df['Seasonal Difference']=df['Close_GSPC']-df['Close_GSPC'].shift(150)
```

```python
 adf_check(df['Seasonal Difference'].dropna())
```

```python
df['Seasonal First Difference']=df['First Difference']-df['First Difference'].shift(1)
```

```python
adf_check(df['Seasonal First Difference'].dropna())
```
