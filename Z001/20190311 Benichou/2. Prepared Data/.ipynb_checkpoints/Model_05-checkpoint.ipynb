#  ARIMA MODELS (AUTO REGRESSIVE INTEGRATED MOVING AVERAGE)
Ne fonctionne pas bien dans le financial. Surtout pour les gestion de stock/vente du retail

Permet de comprendre les data dans le time series, et faire du forecast. Existe en **seasonal** et en **non seasonal (p,d,q)** en fonction de la seasonability des data ou non.

### ARIMA (p,d,q) => NON SEASONAL

#### AR(p) : Autoregression  
A regression model that utilizes the dependent relations between a current observation and observations over a previous period

#### I(d) : Integrated
Differencing observations (substracting an observation from an observation at the previous time step) in order to make the time series stationary

#### MA(q) : Moving Average
A model that uses the dependency between an observation and a residual error from moving average model applied to lagged observations

A **stationary** data set will allow our model to predict that the mean and the variance will be the same in future periods. => The mean, covariance and the variance are constant. To see it, the **Augmented Dickey-Fuller Test**

If not stationary => transform to stationary in order to estimate in which ARIMA term will be the use => Diiference technics

#### STATIONARY TEST
La null hypothesis est que c'est non stationary. C'est la p.value >0.05 reject the null hypothesis

```python
 from statsmodels.tsa.stattools import adfuller
```

```python
result = adfuller(df['Close_GSPC'].dropna())
```

```python
def adf_check (time_series):
    result=adfuller(time_series)
    print('Augmented Dicky-Fuller Test : ')
    labels=['ADF Test Statistic', 'p-value','# of lags','# of observations used']
    for value,label in zip(result,labels):
        print(label+' : '+str(value))
    if result[1] <= 0.05:
        print ('Strong evidence again null hypothesis')
        print('Reject null hypothesis')
        print ('Data has no unit root and is **stationary**')
    else:
        print ('Weak evidence again null hypothesis')
        print('Failed to reject null hypothesis')
        print ('Data has a unit root and is ** non-stationary**')
        
```

```python
adf_check(df['Close_GSPC'].dropna())
```

#### Difference Test
Le but est de tester plusieurs mthode pour trouver quand le modle redeviendra stationary

```python
df['First Difference']=df['Close_GSPC']-df['Close_GSPC'].shift(1)
```

```python
df['First Difference'].plot(figsize=(16,6)) #semble plat donc stationaire
```

```python
 adf_check(df['First Difference'].dropna())
```

#### Si jamais c'tait pas suffisant

```python
df['Second Difference']=df['First Difference']-df['First Difference'].shift(1)
```

```python
df['First Difference'].plot(figsize=(16,6)) #semble plat donc stationaire
```

```python
 adf_check(df['Second Difference'].dropna())
```
