# MONTE CARLO OPTIMIZATION

```python
stocks=pd.concat([aapl['Adj. Close'],cisco['Adj. Close'],ibm['Adj. Close'],amzn['Adj. Close']],axis=1)
stocks.columns=['aapl','cisco','ibm','amzn']
stocks.head()
```

```python
stocks.pct_change(1).mean() # percent daily change
```

```python
stocks.pct_change(1).corr()
```

#### Première approche logarithmique : Log_return

```python
log_ret = np.log(stocks/stocks.shift(1))
log_ret.head()
```

```python
log_ret.hist(bins=100,figsize=(16,12))
```

```python
log_ret.mean() # Moyene du log
```

```python
log_ret.cov() * 252 # Covariance des columns car daily. 52 ou 12 pour weekly ou monthly
```

```python
np.random.seed(101) # On fixe une fois pour toute le random. 101 est arbitraire et est fixé pour la cell

num_ports = 5000 # nbre de portefolio, qu'on pourra augmenter si on veut
all_weights = np.zeros((num_ports,len(stocks.columns)))
ret_arr = np.zeros(num_ports)
vol_arr = np.zeros(num_ports)
sharpe_arr = np.zeros(num_ports)

for ind in range (num_ports):
    
    # weights
    weights = np.array(np.random.random(4)) # On crée des poids aléatoires
    weights = weights/np.sum(weights) # On normalise pour que ce soit < 1

    # Save the Weights
    all_weights[ind,:] = weights
    
    # Expected Portfolio Return
    ret_arr[ind] = np.sum((log_ret.mean() * weights) * 252)

    # Expected Protfolio variance => Volatility
    vol_arr[ind] =  np.sqrt(np.dot(weights.T,np.dot(log_ret.cov()*252,weights)))

    # Sharpe Ratio
    sharpe_arr[ind] = ret_arr[ind]/vol_arr[ind]
```

```python
sharpe_arr.max()
```

```python
sharpe_arr.argmax()
```

```python
all_weights[1420,:]
```

```python
max_sr_ret = ret_arr[sharpe_arr.argmax()]
max_sr_vol = vol_arr[sharpe_arr.argmax()]
```

```python
plt.figure(figsize=(14,10))
plt.scatter(vol_arr,ret_arr,c=sharpe_arr,cmap='plasma')
plt.colorbar(label='Sharpe Ratio')
plt.xlabel('Volatility')
plt.ylabel('Return')
plt.scatter(max_sr_vol,max_sr_ret,c='red',s=50,edgecolors='black')
```

# ARITHMETIC OPTIMIZATION

```python
def get_ret_vol_sr(weights):
    weights = np.array(weights)
    ret = np.sum(log_ret.mean() * weights) * 252
    vol = np.sqrt(np.dot(weights.T,np.dot(log_ret.cov()*252,weights)))
    sr = ret/vol
    return np.array([ret,vol,sr])
```

```python
def neg_sharpe(weights):
    return get_ret_vol_sr(weights)[2]*-1
```

```python
def check_sum(weights):
    return np.sum(weights) - 1 # Return 0 if the sum of the weights is 1
```

```python
cons = {}
cons = ({'type':'eq','fun':check_sum})
cons
```

```python
bounds = ((0,1),(0,1),(0,1),(0,1))
```

```python
init_guess = [0.25,0.25,0.25,0.25]
```

```python
opt_results = minimize(neg_sharpe,init_guess,method='SLSQP',bounds=bounds,constraints=cons)
```

```python
opt_results.x
```

```python
get_ret_vol_sr(opt_results.x)
```

```python
# Sur la méthode graphique, on a vu que cela allait jusqu'à 0.3
frontier_y = np.linspace(0,0.3,100)
```

```python
 def minimize_volatility (weights) :
        return (get_ret_vol_sr(weights)[1])
```

```python
frontier_volatility=[]
# For each return possible, which is the best volatility
for possible_return in frontier_y:
    cons = ({'type':'eq','fun':check_sum},\
           {'type':'eq','fun':lambda w:get_ret_vol_sr(w)[0]-possible_return})
    result = minimize(minimize_volatility,init_guess,method='SLSQP',bounds=bounds,constraints=cons)
    frontier_volatility.append(result['fun'])
```

```python
plt.figure(figsize=(14,10))
plt.scatter(vol_arr,ret_arr,c=sharpe_arr,cmap='plasma')
plt.colorbar(label='Sharpe Ratio')
plt.xlabel('Volatility')
plt.ylabel('Return')
plt.plot(frontier_volatility,frontier_y,'--',linewidth=3.)
```
