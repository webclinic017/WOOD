```python
import random
```

```python
print("Modèle Bollinger Stochastic Descent #1 de Cliff Benson")

print('Mise de départ de 400 000, valeur à 100')

pool_long = 100000
pool_short = 100000
stock_long = 1000
stock_short = 1000
value = 100
gain = 0
cumul = 0
ratio = 0
p = -1
boucle = 0

print('________________')
print("Initialization")
print('________________')

while boucle < 5:
    n = random.randint(-20,20)
    value = 100 + n
    p = p + 1
    boucle = boucle + 1
    
    if n == 0:
            boucle = boucle + 1
            print ('Phase',p)
            print("Valeur d'achat : ", value)
            print ('Pool_long : ',pool_long)
            print ('stock_long : ',stock_long)
            print ('pool_short : ',pool_short)
            print ('stock_short : ',stock_short)
            print('___________________________')
    
    elif n < 0 and pool_long > 0:
        print ('Phase',p)
        print("Valeur d'achat : ", value)
        print('Long on de ',((pool_long/2) / value),' actions')
        print('Dépense du pool long de ',(pool_long/2))
        stock_long = stock_long + ((pool_long/2) / value)
        pool_long = pool_long - (pool_long/2)
        boucle = 0
        print ('Pool_long : ',pool_long)
        print ('stock_long : ',stock_long)
        print ('pool_short : ',pool_short)
        print ('stock_short : ',stock_short)
        print('___________________________')
    
    elif n > 0 and stock_short > 1 :
            print ('Phase',p)
            print("Valeur d'achat : ", value)
            print('Short de ',(stock_short/2),' actions')
            print('Gain du pool short de ',((stock_short/2) * value))
            pool_short = pool_short + ((stock_short/2) * value)
            stock_short = stock_short - (stock_short/2)
            boucle = 0
            print ('Pool_long : ',pool_long)
            print ('stock_long : ',stock_long)
            print ('pool_short : ',pool_short)
            print ('stock_short : ',stock_short)
            print('___________________________')
            
    else :
        boucle = boucle + 1
        print ('Phase',p)
        print("Valeur d'achat : ", value)
        print ('Pool_long : ',pool_long)
        print ('stock_long : ',stock_long)
        print ('pool_short : ',pool_short)
        print ('stock_short : ',stock_short)
        print('___________________________')
        if boucle == 5:
            exit
print('___________________________________________________________________')
print("Il n'y a plus de fond à dépenser. Termination")
print('___________________________________________________________________')

    
print ('Phase',p)
print("Valeur d'achat : ", value)
print ('Pool_long : ',pool_long)
print ('stock_long : ',stock_long)
print ('pool_short : ',pool_short)
print ('stock_short : ',stock_short)
print('___________________________')
    
    

pool_short = pool_short + (stock_short * value)
stock_short = 0
pool_long = pool_long + (stock_long * value)
stock_long = 0

print('___________________________________________________________________')
print("Il n'y a plus de stock à vendre. Finalization")
print('___________________________________________________________________')

print ('Phase finale')
print("Valeur d'achat : ", value)
print ('Pool_long : ',pool_long)
print ('stock_long : ',stock_long)
print ('pool_short : ',pool_short)
print ('stock_short : ',stock_short)
print('___________________________')

cumul = pool_short + pool_long
print('le cumul est de : ',cumul)
gain = cumul - 400000
print ("Le gain total est de : ",gain)
ratio = gain/4000
print('Le ratio est de ',ratio,'%')
```
