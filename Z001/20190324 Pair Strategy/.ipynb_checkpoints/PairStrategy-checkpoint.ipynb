```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
%matplotlib inline
import seaborn as sns
sns.set(palette='bright',style='whitegrid')
import numpy as np
import pandas as pd
import quandl
import pandas_datareader as web
from statsmodels.tsa.stattools import coint
import statsmodels
from statsmodels.tsa.stattools import adfuller
import itertools
import scipy
from scipy import stats
import time
from matplotlib.backends.backend_pdf import PdfPages
```

```python
constituents = pd.read_csv('constituents_csv.csv')
constituents.head()
```

```python
symbols = constituents['Symbol']
symbols.head()
```

```python
symbols.head()
```

```python
lenght = len(symbols)
lenght
```

```python
start = pd.to_datetime('2014-03-19')
end = pd.to_datetime('2019-03-19')
```

```python
symbols_list = symbols.to_list()
```

```python
# pairs = [(symbols_list[i], symbols_list[i+1]) for i in range(len(symbols_list)-1) if i%2==0]
pairs = (list(itertools.permutations(symbols_list,2)))
lenght = len(pairs)
selection=[]
```

```python
for i in range (0,len(symbols_list)):
    try:
        label = symbols_list[i]
        pair = web.DataReader(label,'yahoo',start,end)
        X = pair['Adj Close']
        adf = adfuller(X)
    
        if adfuller(X)[1]<=adfuller(X)[4]['5%'] == True:
            selection.append(X)
            print('\n ',label)
    except :
        pass
```

```python
print(selection)
```

```python
len(selection)
```

```python
i = 0
label_0 = pairs[i][0]
label_1 = pairs[i][1]

pair_0 = web.DataReader(label_0,'yahoo',start,end)
pair_1 = web.DataReader(label_1,'yahoo',start,end)
spy_etf = web.DataReader('SPY','yahoo',start,end)
def zscore(series):
    return (series - series.mean()) / np.std(series)

beta,alpha,r_value,p_value,std_err = stats.linregress(pair_0['Adj Close'].iloc[1:],\
                                                                  pair_1['Adj Close'].iloc[1:])
print('__________________________________________')
print('Résultat pour la paire :',label_0,', ',label_1)
print('__________________________________________')
print('Bêta : ',beta)
print('Alpha : ',alpha)
print('r_value : ',r_value)
print('p_value : ',p_value)
print('std_err : ',std_err)
print('__________________________________________')
print('\n'*2)

X = pair_0['Adj Close']
Y = pair_1['Adj Close']
score, pvalue, _ = coint(X,Y)
if pvalue <= 0.05:
    print('Les paires ',label_0,' et ',label_1,' sont cointégrées. La p.value est de :',pvalue,\
                      ' et le score de : ',score)
else:
    print('Les paires ',label_0,' et ',label_1,' ne sont pas cointégrées. La p.value est de :',pvalue,\
                      ' et le score de : ',score)
spread = X - Y
```

```python
plt.figure()
pair_0['Adj Close'].plot(label=label_0,figsize=(16,6),title='Comparaison des indices')
pair_1['Adj Close'].plot(label=label_1,figsize=(16,6))
spy_etf['Adj Close'].plot(label='SPY Index',figsize=(16,6))
plt.legend()

print('Spread : ',label_0,' moins ',label_1)
plt.figure()
spread.name = 'spread'
zscore(spread).plot(figsize=(16,6))
plt.axhline(zscore(spread).mean(), color='black')
plt.axhline(1.0, color='red', linestyle='--')
plt.axhline(-1.0, color='green', linestyle='--')
plt.legend(title='Zscore : (spread-spread.mean())/np.std(spread)')
print('\nSimple Strategy: \nGo "Long" the spread whenever the z-score is below -1.0.\
\nGo "Short" the spread when the z-score is above 1.0.\nExit positions when the z-score approaches zero\n')
```
